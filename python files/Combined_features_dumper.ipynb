```python
import pickle
import pandas as pd
import numpy as np
import os
from scipy.sparse import hstack
from IPython.display import display

pd.set_option('display.max_rows', 1000)
pd.set_option('max_colwidth',60)

table_matrix = pd.DataFrame.from_dict(pickle.load(open('metric_table_brown.pickle', 'rb')))
    
table_matrix = table_matrix[['Feature', 'Classifier', 'Accuracy', 'Precision', 'Recall', 'F1 score', 'Confusion Matrix']]
a = table_matrix['Confusion Matrix']
# table_matrix['True Count'] = list(map(lambda x: x[0][0] + x[1][1] + x[2][2], a))
def feature_type(stringy):
    splity = stringy.split('_')
    if splity[-1] == 'alphabet':
        return 'punctuation'
    if splity[-1] == 'pos':
        return 'pos'
    if splity[0] == 'capitals':
        return 'capital'
    if 'char' in splity:
        return 'character'
    return 'word'
table_matrix['Feature type'] = list(map(feature_type, table_matrix['Feature']))
del a
print(table_matrix['Feature type'].value_counts())
display(table_matrix)
```

# Top char n-grams

```python
top_number = 4
evaluation_critera = 'F1 score'
```

```python
top_char_features = table_matrix[(table_matrix['Feature type']=='character')].sort_values(evaluation_critera, ascending=False).head(top_number)
top_char_features
```

# Top word n-grams

```python
top_word_features = table_matrix[(table_matrix['Feature type']=='word')].sort_values(evaluation_critera, ascending=False).head(top_number)
display(top_word_features)
```

# Top punctuation n-grams

```python
top_punctuation_features = table_matrix[(table_matrix['Feature type']=='punctuation')].sort_values(evaluation_critera, ascending=False).head(top_number)
display(top_punctuation_features)
```

# top capital n-grams

```python
top_capital_features = table_matrix[(table_matrix['Feature type']=='capital')].sort_values(evaluation_critera, ascending=False).head(3)
display(top_capital_features)
```

# top pos n-grams

```python
top_pos_features = table_matrix[(table_matrix['Classifier']=='SVM_linear') & (table_matrix['Feature type']=='pos')].sort_values(evaluation_critera, ascending=False).head(top_number)
display(top_pos_features)

if top_pos_features.empty:
    top_pos_features = {'Feature': 
                        ['fivegram_pos',
                        '2-3_gram_pos',
                        '1-3_gram_pos',
                        '2-4_gram_pos',
                        '3-4_gram_pos']}
    print("\n\n", top_pos_features)
```

```python
top_feature_list_to_path = [
                    ['features/' + i + '.pickle' for i in list(top_char_features['Feature'])],
                    ['features/' + i + '.pickle' for i in list(top_word_features['Feature'])],
                    ['extra_features/' + i + '.pickle' for i in list(top_punctuation_features['Feature'])],
                    ['extra_features/' + i + '.pickle' for i in list(top_capital_features['Feature'])]
#                     ['pos_features/' + i + '.pickle' for i in list(top_pos_features['Feature'])]
                ]

for i in top_feature_list_to_path:
    for j in i:
        if not os.path.isfile(j):
            print(str("\nERROR: " + i + " FILE NOT FOUND\n"))
            continue

print(top_feature_list_to_path)
```

```python
feature_dictionary = {}

feature_from_path = lambda x: x.split('/')[-1].split('.')[0]

for i in top_feature_list_to_path:
    for j in i:
        feature_dictionary[feature_from_path(j)] = pickle.load(open(j, 'rb'))
    
feature_dictionary_printer = lambda x :  print("X_train: {}\ny_train: {}\nX_test: {}\ny_test: {}\n\n".format(
                                                x['X_train'].shape,
                                                x['y_train'].shape,
                                                x['X_test'].shape,
                                                x['y_test'].shape
                                        ))

for i in feature_dictionary:
    print(i)
    feature_dictionary_printer(feature_dictionary[feature_from_path(i)])
```

```python
while len(top_feature_list_to_path) < 4:
    top_feature_list_to_path.append([])
```

```python
%%time

from itertools import combinations

listy = [['a','b', 'c'], [1, 2, 3], ['+', '-', '*']]#, ['Y', 'Z']]
# listy = [['a','b'], [1, 2], ['+', '-']]
# listy = [['a'], [1], ['+']]
# print()


def combinatrics_ultimatrix(listy):
    tempo = []
    for i in listy:
        if type(i) == list:
            tempo += i
        else:
            tempo.append(i)
    combo_dict = {
        'Feature 1': [],
        'Feature 2': [],
        'Feature 3': [],
        'Feature 4': [],
        'Data': []
    }
    for num_elements in range(2, len(tempo)+1):
        for combo in combinations(tempo, num_elements):
            combo_set = set(combo)
#             print(combo_set)
            this_set_is_doable = True
            for temp_combo in list(map(lambda x: list(combinations(x, 2)), listy)):
                for temp_temp_combo in temp_combo:
                    if set(temp_temp_combo).issubset(combo_set):
                        this_set_is_doable = False
                        break
                if not this_set_is_doable:
                    break
            if not this_set_is_doable:
                continue
            
            #if code reaches here, means we have a workable combination
            combo_set = list(combo_set)
            for i in range(1, 5):
                if len(combo_set) >= i:
                    combo_dict['Feature ' + str(i)].append(feature_from_path(combo_set[i-1]))
                else:
                    combo_dict['Feature ' + str(i)].append('-')
                    
            combo_dict['Data'].append({
                'X_train': hstack(list(map(lambda x: feature_dictionary[feature_from_path(x)]['X_train'], combo_set))),
                'y_train': feature_dictionary[feature_from_path(combo_set[0])]['y_train'],
                'X_test': hstack(list(map(lambda x: feature_dictionary[feature_from_path(x)]['X_test'], combo_set))),
                'y_test': feature_dictionary[feature_from_path(combo_set[0])]['y_test']
            })
#             train_data = hstack(list(map(lambda x: feature_dictionary[x]['X_train'], combo_set)))
#             test_data = hstack(list(map(lambda x: feature_dictionary[x]['X_test'], combo_set)))
            
#             print(train_data.shape, test_data.shape)
        print("Completed ", num_elements)
        print()
    
    return combo_dict
    
resulto = combinatrics_ultimatrix(top_feature_list_to_path) 
print("Number of feature combinations:", len(resulto['Data']))
#PRINT TABLE_LIKE FORMAT
# for i in range(len(resulto['Feature 1'])):
#     for j in range(1, 5):
#         print(resulto['Feature '+str(j)][i], end='\t\t')
#     print()
     
```

```python
########for printing
for i in range(len(resulto['Data'])):
    for k in range(1, 5):
        if resulto['Feature '+str(k)][i] == '-':
            continue
        print("{}\n".format(resulto['Feature '+str(k)][i]))
        feature_dictionary_printer(feature_dictionary[resulto['Feature '+str(k)][i]])
    print("{}\n{}\n{}\n{}\n".format(resulto['Feature 1'][i], resulto['Feature 2'][i], 
                                    resulto['Feature 3'][i], resulto['Feature 4'][i]))
    feature_dictionary_printer(resulto['Data'][i])
```

```python
%%time
parts = 10
folder = 'combined_features'
part_length = len(resulto['Data']) // parts

for i in list(filter(lambda x: x.split('.')[-1] == 'pickle', os.listdir(folder))):
    os.remove(os.path.join(folder, i))
    print("Removed", i)
    
for i in range(parts+1):
    temp_dict = {}
    for key in resulto:
        temp_dict[key] = resulto[key][i*part_length: (i+1)*part_length]
    file_name = 'part_features_' + str(i) + '.pickle'
    pickle.dump(temp_dict, open(os.path.join(folder, file_name), 'wb'))
    print("Dumped ", file_name)

#     pickle.dump(resulto, open('combined_features/all_features.pickle', 'wb'))
```
