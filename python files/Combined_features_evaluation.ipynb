```python
import pickle
import pandas as pd
import numpy as np
import os
from IPython.display import display

#set pandas table parameters
pd.set_option('display.max_rows', 1000)
pd.set_option('max_colwidth',70)
```

```python
count = 0
while os.path.exists(os.path.join('combined_features', 'folder_'+str(count))):
    count += 1
read_folder = os.path.join('combined_features', 'folder_'+str(count-1))
# read_folder = os.path.join('combined_features', 'folder_'+str(7))

path_to_svm_files = list(map(lambda x: os.path.join(read_folder, x), os.listdir(read_folder)))
print(read_folder, ':', len(path_to_svm_files), "files found.")

evaluation_dict = pickle.load(open(path_to_svm_files[0], 'rb'))

for label in evaluation_dict:
    evaluation_dict[label] = [evaluation_dict[label]]

for i in range(1, len(path_to_svm_files)):
    data = pickle.load(open(path_to_svm_files[i], 'rb'))
    for label in data:
        evaluation_dict[label].append(data[label])

table = pd.DataFrame.from_dict(evaluation_dict)
table = table[['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4', 'Accuracy', 'Recall', 'Precision', 'F1 score', 'Confusion Matrix']]
del table['Precision']
del table['Recall']
display(table.sort_values('F1 score', ascending=False))
# display(table.sort_values('F1 score', ascending=False))
```
